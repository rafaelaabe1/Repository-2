```python
# Projeto caixa eletrônico

# Você foi contratado para desenvolver um sistema simples de caixa eletrônico.

# O sistema deverá funcionar através de um menu interativo utilizando "while»

# O programa deve iniciar com um saldo inicial de R$1000,00 e permitir que o usuário escolha diferentes operações até decidir sair do sistema.

# Ver saldo
# Depositar
# Sacar Dinheiro
# Sair

# Regras:

# Quando depositar somar ao saldo anterior;
# Quando Sacar subtrair do saldo anterior
# Mostrar o saldo para o usuário

# Inicializa o saldo com R$ 1000,00

print("-----MENU-----")
print("")
print("1 - ver saldo")
print("2 - depositar")
print("3 - sacar")
print("4 - sair")
print("-------------")
print("")

saldo = 1000
opcao = ""

while opcao != 0:
    try:

      opcao = int(input("escolha uma opção de 1 a 4: "))

      if opcao == 1:
        print (f"O seu saldo atual: {saldo}")

      elif opcao == 2:
        deposito = float (input("Digite o valor que deseja deposita:"))
        saldo = saldo + deposito

      elif opcao == 3:
        deposito = float(input("Digite o valor que deseja sacar:"))
        saldo = saldo - deposito
        print ("")

    except:
      print("escolha uma opção de 1 a 4 para sair: ")
```

```python
# fazer um jogo de adivinhação

# O usúario irá adivinhar um número de 0 até 10.

# Regras

# O usúario terá 03 chances para adivinhar
# Informar se o palpite do usúario é maior ou menor que o número.
# Imprimir mensagem de está quente, morno ou frio.

import random

ns = random.randint(0, 10)
chances = 3

print("Bem vindo ao jogo de adivinhação")
print("escolha um número entre 0 e 10.")
print(f"Você tem {chances} chances para adivinhar.")

for tentativa in range(chances):
    try:
        palpite = int(input(f"Tentativa {tentativa + 1}/{chances}: Digite: "))

        if palpite == ns:
            print(f"Você adivinhou o número {n_s} em {tentativa + 1} tentativas")
            break
        elif palpite < ns:
            print("È muito baixo.")
        else:
            print("È muito alto.")

        diferenca = abs(ns - palpite)
        if diferenca <= 1:
            print("Esta super quente")
        elif diferenca <= 3:
            print("Esta quente")
        elif diferenca <= 5:
            print("Esta morno")
        else:
            print("Esta frio")

    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

else:
    print(f"Suas {chances} chances acabaram. O número secreto era {ns}.")
print("Fim do jogo.")
```
